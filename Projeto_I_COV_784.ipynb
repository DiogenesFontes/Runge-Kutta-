**Considere a seguinte ODE linear, cuja variável independente é t e cuja variável dependente é w(t),**

**dω/dt = λω     (1)**

**Considere, ainda: λ=-0.97736 e w(0)=1. Pergunta-se**

**a) Plotar a solução analítica exata deste problema até o instante t=Tmax=2 ?**


```python
def modelo (ω,t):
    λ = -0.97736
    dωdt = λ*ω
    return   dωdt

ω0 = 1
t = np.linspace(0,2)
ω = odeint(modelo,ω0,t)


f = plt.figure()
f.set_figwidth(10)
f.set_figheight(8)
plt.plot(t,ω,'--o', label = 'Analítica')
plt.legend(loc="upper right")
plt.title("SOLUÇÃO ANALÍTICA", fontsize = 14)
plt.xlabel('tempo', fontsize=14)
plt.ylabel('ω', fontsize=14)
plt.margins(x=0)
plt.margins(y=0)
plt.xlabel("Tempo", fontsize = 15)
plt.ylabel("ω", fontsize = 15)
plt.tick_params(axis='both', labelsize = 15)
plt.show()
```

**d) Repita o item (b) utilizando um método de Runge-Kutta de 2a ordem e Npassos=10.**

```python
def modelo(ω, t):
    λ = -0.97736
    dωdt = λ * ω
    return dωdt


ω0 = 1
t = np.linspace(0, 2)
ω = odeint(modelo, ω0, t)

def f(x, y):
    r = -0.97736
    return r * y
    # return (e ** r * y) * (r * y)


x0 = 0
y0 = 1

a = 0
b = 1
t_max = 2
N: int = (10 + 1)
h = t_max / N

xi = [x0]
yi = [y0]

for i in range(N):
    k1 = f(x0, y0)
    k2 = f(x0 + h, y0 + h * k1)
    yk = y0 + 0.5 * (k1 + k2) * h
    x0 = x0 + h
    y0 = yk

    print(f"({round(x0, 2)}, {round(y0, 2)})")

    xi.append(round(x0, 2))
    yi.append(round(yk, 2))


f = plt.figure()
f.set_figwidth(10)
f.set_figheight(8)
plt.plot(t, ω, '--o', label = 'Analítica')
plt.plot(xi,yi, '--', color = 'red', label = 'Runge Kutta 2 Npassos=10')
plt.legend(loc="upper right")
plt.margins(x=0)
plt.margins(y=0)
plt.xlabel("Tempo", fontsize = 15)
plt.ylabel("ω", fontsize = 15)
plt.tick_params(axis='both', labelsize = 15)
```

**Repita o item (b) utilizando um método de Runge-Kutta de 1a ordem e Npassos=20.**

```python
def modelo(ω, t):
    λ = -0.97736
    dωdt = λ * ω
    return dωdt


ω0 = 1
t = np.linspace(0, 2)
ω = odeint(modelo, ω0, t)

def f(u, v):
    r = -0.97736
    return r * v



u0 = 0
v0 = 1

a = 0
b = 1
t_max = 2
N: int = (20)
h = t_max / N

ui = [u0]
vi = [v0]
r = -0.97736

for i in range(N):
    k1 = f(u0, v0) + h * (e ** r * u0)
    u0 += h
    v0 += k1 * h

    print(f"({round(u0, 2)}, {round(v0, 2)})")

    ui.append(round(u0, 2))
    vi.append(round(v0, 2))

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(8)
plt.plot(t, ω, '--o', label = 'Analítica')
plt.plot(ui,vi, '-', color = 'red', label = 'Runge Kutta 1 Npassos=20')
plt.legend(loc="upper right")
plt.margins(x=0)
plt.margins(y=0)
plt.xlabel("Tempo", fontsize = 15)
plt.ylabel("ω", fontsize = 15)
plt.tick_params(axis='both', labelsize = 15)

```
